### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
# Display the first few rows
print("🔹 First few rows of the dataset:")
display(df.head())

# Display column names
print("\n🔹 Column names:")
print(df.columns.tolist())

# Display data types and non-null counts
print("\n🔹 Data info:")
df.info()

# Display dataset shape
print(f"\n🔹 Dataset shape: {df.shape[0]} rows × {df.shape[1]} columns")

# Display basic statistics for numeric columns
print("\n🔹 Summary statistics:")
display(df.describe())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
# STEP 3.1: Check for missing values
print("🔹 Missing values per column:\n")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0].sort_values(ascending=False))

# Optional: View percentage of missing data
print("\n🔹 Percentage of missing values:\n")
percent_missing = (df.isnull().mean() * 100).round(2)
print(percent_missing[percent_missing > 0].sort_values(ascending=False))
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
# STEP 4.1: Value counts for Employment
print("🔹 Employment Distribution:\n")
print(df['Employment'].value_counts(dropna=False))
print("\n" + "-"*50)

# STEP 4.2: Value counts for Job Satisfaction
print("🔹 Job Satisfaction Distribution:\n")
print(df['JobSat'].value_counts(dropna=False))
print("\n" + "-"*50)

# STEP 4.3: Value counts for Years of Coding Experience
print("🔹 Professional Coding Experience Distribution:\n")
print(df['YearsCodePro'].value_counts(dropna=False).sort_index())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
# Count JobSat values
job_sat_counts = df['JobSat'].value_counts()

# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(job_sat_counts, labels=job_sat_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Job Satisfaction Distribution')
plt.axis('equal')  # Equal aspect ratio makes the pie circular
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
from collections import Counter

# Helper: Split and count all language mentions
def get_language_counts(series):
    return Counter(lang.strip() for entry in series.dropna() for lang in entry.split(';'))

# Get counts
worked_with_counts = get_language_counts(df['LanguageHaveWorkedWith'])
want_to_work_with_counts = get_language_counts(df['LanguageWantToWorkWith'])

# Convert to DataFrame for comparison
lang_df = pd.DataFrame({
    'WorkedWith': pd.Series(worked_with_counts),
    'WantToWorkWith': pd.Series(want_to_work_with_counts)
}).fillna(0).astype(int)

# Sort by popularity
lang_df = lang_df.sort_values(by='WorkedWith', ascending=False)

# Preview
lang_df.head()
```

```python
# Plot Top 10 Most Common Languages
top_n = 10
top_langs = lang_df.head(top_n)

top_langs.plot(kind='bar', figsize=(10, 6), color=['#1f77b4', '#ff7f0e'])
plt.title(f'Top {top_n} Programming Languages: Worked With vs. Want To Work With')
plt.ylabel('Count')
plt.xlabel('Programming Language')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
# Step 1: Cross-tabulate RemoteWork by Country
remote_by_region = pd.crosstab(df['Country'], df['RemoteWork'])

# Step 2: Sort by total responses (optional: to focus on top 10)
top_regions = remote_by_region.sum(axis=1).sort_values(ascending=False).head(10).index
remote_top = remote_by_region.loc[top_regions]

# Step 3: Plot grouped bar chart
remote_top.plot(kind='bar', figsize=(12, 6))
plt.title('Remote Work Distribution by Region (Top 10)')
plt.xlabel('Country / Region')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Remote Work Type')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
# Step 1: Convert JobSat to numeric scale
job_sat_map = {
    'Very dissatisfied': 1,
    'Dissatisfied': 2,
    'Neither satisfied nor dissatisfied': 3,
    'Satisfied': 4,
    'Very satisfied': 5
}
df['JobSatScore'] = df['JobSat'].map(job_sat_map)

# Step 2: Clean YearsCodePro
def convert_experience(x):
    if pd.isna(x):
        return None
    if x == 'Less than 1 year':
        return 0.5
    elif x == 'More than 50 years':
        return 51
    try:
        return float(x)
    except:
        return None

df['YearsCodeProNum'] = df['YearsCodePro'].apply(convert_experience)

# Step 3: Drop rows with missing values in either column
cor_df = df[['JobSatScore', 'YearsCodeProNum']].dropna()

# Step 4: Calculate correlation
pearson_corr = cor_df.corr(method='pearson').iloc[0, 1]
spearman_corr = cor_df.corr(method='spearman').iloc[0, 1]

print(f"📈 Pearson Correlation: {pearson_corr:.3f}")
print(f"📈 Spearman Correlation: {spearman_corr:.3f}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
# Step 1: Cross-tabulate Employment vs. Education Level
edu_emp_ct = pd.crosstab(df['EdLevel'], df['Employment'])

# Optional: Focus on most common education levels (to keep it readable)
top_ed_levels = edu_emp_ct.sum(axis=1).sort_values(ascending=False).head(6).index
edu_emp_ct = edu_emp_ct.loc[top_ed_levels]

# Step 2: Plot stacked bar chart
edu_emp_ct.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20c')
plt.title('Employment Type by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
